## $D+(YRI, NA19664, Archaic, EPO)$

```python
# Intialize a dictionary to store the genotype matricies.
mxl_arc_gts_742kb = {}
mxl_arc_gts_72kb = {}
# For every archaic.
for arc in ['DEN', 'ALT', 'CHA', 'VIN']:
    # Load the genotype matricies.
    gt_742kb, pos_742kb = af.load_hap_region(f'tgp_{arc.lower()}_masked_aa', 12, 40272001, 41014000)
    gt_72kb, pos_72kb = af.load_hap_region(f'tgp_{arc.lower()}_masked_aa', 12, 40759001, 40831000)
    # Update the dictionaries.
    mxl_arc_gts_742kb[arc] = gt_742kb
    mxl_arc_gts_72kb[arc] = gt_72kb
```

### 742kb

```python
# Load the observed values.
mxl_dplus_742kb_obs = af.build_mxl_arc_scenario_obs_dicc(mxl_arc_gts_742kb)
# Load the windowed values.
mxl_dplus_742kb_winds = af.build_mxl_arc_scenario_window_dicc(742)
# Compile the summaries.
mxl_dplus_742kb_summary = af.compile_mxl_arc_dplus_scenario_summary(
    mxl_dplus_742kb_obs, mxl_dplus_742kb_winds, 742,
)
```

```python
# Show the results.
mxl_dplus_742kb_summary
```

### 72kb

```python
# Load the observed values.
mxl_dplus_72kb_obs = af.build_mxl_arc_scenario_obs_dicc(mxl_arc_gts_72kb)
# Load the windowed values.
mxl_dplus_72kb_winds = af.build_mxl_arc_scenario_window_dicc(72)
# Compile the summaries.
mxl_dplus_72kb_summary = af.compile_mxl_arc_dplus_scenario_summary(
    mxl_dplus_72kb_obs, mxl_dplus_72kb_winds, 72,
)
```

```python
# Show the results.
mxl_dplus_72kb_summary
```

### Plots

```python
# Plot the results for the 742kb region.
af.plot_mxl_arc_dplus(mxl_dplus_742kb_obs, mxl_dplus_742kb_winds, 742)
```

```python
# Plot the results for the 72kb region.
af.plot_mxl_arc_dplus(mxl_dplus_72kb_obs, mxl_dplus_72kb_winds, 72)
```
