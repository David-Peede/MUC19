```python
# Import packages.
import analysis_functions as af
import pandas as pd
# Print version numbers.
print('pandas', pd.__version__)
```

```python
# Intialize my pandas preferences.
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Load the genotype matricies.
tgp_72kb_gt, _ = af.load_hap_region('tgp_mod_arc_anc', 12, 40758000, 40830000)
arc_72kb_gt, _ = af.load_hap_region('arc_anc', 12, 40758000, 40830000)
```

## ((YRI, OOA), Denisovan)

```python
# Intialize sample configuration lists.
p1s = ['YRI']
p2s = [
    'PEL', 'MXL', 'CLM', 'PUR', # AMR.
    'BEB', 'STU', 'ITU', 'PJL', 'GIH', # SAS.
    'CHB', 'KHV', 'CHS', 'JPT', 'CDX', # EAS.    
    'TSI', 'CEU', 'IBS', 'GBR', 'FIN', # EUR.
]
p3s = ['DEN']
# Load the windowed results.
yri_ooa_den_winds = af.build_tgp_arc_scenario_window_dicc('tgp', p1s, p2s, p3s, 72)
# Calculate the observed results.
yri_ooa_den_obs = af.calculate_tgp_arc_scenarios(tgp_72kb_gt, yri_ooa_den_winds, p1s, p2s, p3s, 72)
```

```python
# Compile the results summary.
af.build_tgp_arc_scenario_tables(yri_ooa_den_obs, yri_ooa_den_winds, p1s, p2s, p3s, 72)
```

```python
# Plot the results.
af.plot_tgp_arc_scenarios('YRI', 'DEN', yri_ooa_den_winds, yri_ooa_den_obs, r'$MUC19$', 72)
```
