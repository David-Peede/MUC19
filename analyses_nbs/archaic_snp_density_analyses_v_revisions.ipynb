# Archaic SNP Denisty Analyses

```python
# Import packages.
import analysis_functions as af
```

```python
# Load the genotype matricies.
tgp_arcs_no_aa_742kb_gt, _ = af.load_hap_region('tgp_arcs_masked_no_aa', 12, 40272001, 41014000)
tgp_arcs_no_aa_72kb_gt, _ = af.load_hap_region('tgp_arcs_masked_no_aa', 12, 40759001, 40831000)
```

## 742kb

```python
# Load the observed values.
tgp_arc_snps_742kb_obs = af.tgp_arc_snp_density(tgp_arcs_no_aa_742kb_gt)
# Load the windowed values.
tgp_arc_snps_742kb_winds = af.load_tgp_arc_snp_density_windows(742)
# Compile the summaries.
tgp_den_snps_742kb_summary = af.compile_tgp_arc_snp_denisty_summary(
    tgp_arc_snps_742kb_obs, tgp_arc_snps_742kb_winds, 'DEN', 742,
)
tgp_nea_snps_742kb_summary = af.compile_tgp_arc_snp_denisty_summary(
    tgp_arc_snps_742kb_obs, tgp_arc_snps_742kb_winds, 'NEA', 742,
)
```

```python
# Show the Denisovan-specific snp density results.
tgp_den_snps_742kb_summary
```

```python
# Show the Neanderthal-specific snp density results.
tgp_nea_snps_742kb_summary
```

## 72kb

```python
# Load the observed values.
tgp_arc_snps_72kb_obs = af.tgp_arc_snp_density(tgp_arcs_no_aa_72kb_gt)
# Load the windowed values.
tgp_arc_snps_72kb_winds = af.load_tgp_arc_snp_density_windows(72)
# Compile the summaries.
tgp_den_snps_72kb_summary = af.compile_tgp_arc_snp_denisty_summary(
    tgp_arc_snps_72kb_obs, tgp_arc_snps_72kb_winds, 'DEN', 72,
)
tgp_nea_snps_72kb_summary = af.compile_tgp_arc_snp_denisty_summary(
    tgp_arc_snps_72kb_obs, tgp_arc_snps_72kb_winds, 'NEA', 72,
)
```

```python
# Show the denisovan-specific snp density results.
tgp_den_snps_72kb_summary
```

```python
# Show the neanderthal-specific snp density results.
tgp_nea_snps_72kb_summary
```
