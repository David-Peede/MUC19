## 742kb

```python
# Load the observed values.
tgp_ihs_742kb_obs = af.load_focal_region_ihs(742)
# Load the windowed values.
tgp_ihs_742kb_winds = af.load_ihs_windows(742)
# Compile the summary.
tgp_ihs_742kb_summary = af.compile_ihs_window_summary(tgp_ihs_742kb_obs, tgp_ihs_742kb_winds, 742)
```

```python
# Show the proportion of snps with |iHS| > 2 results.
tgp_ihs_742kb_summary
```

```python
# Demonstrate that MXL is the only population with a cluster of |iHS| > 2 in the top 1% of the windowed distribution.
tgp_ihs_742kb_summary[tgp_ihs_742kb_summary['Focal 742kb Region is in Top 1%?'].values]
```
