## 72kb

```python
# Intialize a dictionary to store the archaic genotype matricies.
arc_gts_72kb = {}
# For every archaic.
for arc in ['DEN', 'ALT', 'CHA', 'VIN']:
    # Load the genotype matrix.
    arc_gt, _ = af.load_hap_region(f'{arc.lower()}_masked_no_aa', 12, 40759001, 40831000)
    # Update the dictionary.
    arc_gts_72kb[arc] = arc_gt
# Load the tgp data.
tgp_mod_no_aa_72kb_gt, _ = af.load_hap_region('tgp_mod_no_aa', 12, 40759001, 40831000)
```

```python
# Load the observed values.
arc_het_72kb_obs = af.arc_obs_het(arc_gts_72kb)
# Load the windowed values.
arc_het_72kb_winds = af.load_arc_het_windows(72)
# Compile the summary.
arc_het_72kb_summary = af.compile_arc_het_summary(arc_het_72kb_obs, arc_het_72kb_winds, 72)
```

```python
# Show the archaic summary.
arc_het_72kb_summary
```

```python
# Load the observed values.
tgp_het_72kb_obs = af.tgp_obs_het(tgp_mod_no_aa_72kb_gt)
# Load the windowed values.
tgp_het_72kb_winds = af.load_tgp_het_windows(72)
# Compile the summary.
tgp_het_72kb_summary = af.compile_tgp_het_summary(tgp_het_72kb_obs, tgp_het_72kb_winds, 72)
```

```python
# Show the focal group summary.
tgp_het_72kb_summary
```
